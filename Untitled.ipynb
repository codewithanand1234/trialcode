```python

# x=10
# y=6
# mse mean square error
#[y[i]-ypred[i]**2]/n
# (y[i]-(w*xi+b))^2/n
# d/dw=-2/n*(xi)(y[i]-(yp))
# d/db=2/n(y[i]-yp)

# lr=0.001
# w=0 and b=0
# w=w-lr*dw
# b=b-lr*db
import numpy as np
def gradient(x,y):
    cw=0
    cb=0
    iter1=5000
    lr=0.001
    n=len(x)
    for i in range(iter1):
        yp=cw*x+cb
        dw=-2/n*(sum(x*(y-yp)))
        cost=1/n*(sum([val**2 for val in (y-yp)]))
        db=2/n*(sum(y-yp))
        cw=cw-lr*dw
        cb=cb-lr*db
        print(i,cw,cb,cost)
x=np.array([1,2,3,4,5])
y=np.array([1,3,5,7,9])                  
gradient(x,y)                  
        
        


```
